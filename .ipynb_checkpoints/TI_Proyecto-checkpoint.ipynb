```python
#Bibliotecas para la ejecución
import random, copy
import matplotlib.pyplot as plt

from math import floor, ceil
```

```python
#Cargando los conjuntos de entrenamiento y prueba
train = open('Datos_Entrenamiento.txt','r')
test = open('Datos_Prueba.txt','r')

#Strip. Remueve espacios al inicio o final de la cadena.
#Split(','). Separa palabras cada que se encuentre una coma.
#map. Mapea los datos de la lista de string a Flotantes.
entrenamiento = [list(map(float, (linea.strip()).split(','))) for linea in train]
prueba = [list(map(float, (linea.strip()).split(','))) for linea in test]

#Vizualizando tamaño de los conjuntos.
print("Entrenamiento: {} \nPrueba: {}".format(len(entrenamiento), len(prueba)))

#Cerrando los archivos.
train.close()
test.close()

#Contando las clases 
#Mínimo 2 clases
clases_totales = [0, 1]
for individuo in entrenamiento:
    if individuo[len(individuo)-1] not in clases_totales:
        clases_totales.append(int(individuo[len(individuo)-1]) )
print("Clases en el conjunto de entrenamiento: ", len(clases_totales))
```

## Distancia Euclidiana

Donde P = (p1, p2, p3, ... , pn) y Q = (p1, p2, p3, ... , pn) son vectores de características.

## Distancia Manhattan

Donde P = (p1, p2, p3, ... , pn) y Q = (p1, p2, p3, ... , pn) son vectores de características.

## Distancia Minkowski 

Para p = 1, la expresión anterior coincide con la distancia Manhattan, y para p = 2, coincide con la distancia Euclidea. Donde X = (x1, x2, x3, ... , xn) y Y = (y1, y2, y3, ... , yn) son vectores de características.

```python
#Funciones de distancia
#Exponente = 1 ---> Distancia Manhattan 
#Exponente = 2 ---> Distancia Euclidiana
#exponente >=3 ---> Distancia Minkowski

def funciones(individuo, nuevo_individuo, caracteristicas, exponente):
    distancia = 0.0
    for caracteristica in range(caracteristicas):
        distancia += abs(individuo[caracteristica] - nuevo_individuo[caracteristica])**exponente
    return( (distancia**(1/exponente)))
```

```python
#Visualización de los datos
def visualizacion(entrenamiento, caracteristicas, iteracion):
    #Definición de los colores a usar
    colores = {
        0: 'red',
        1: 'olive',
        2: 'darkorange',
        3: 'lime',
        4: 'aqua',
        5: 'blue',
        6: 'fuchsia'}
    
    #Tamaño de la figura
    fig = plt.figure(figsize=(8, 8))
    #Figura 3D
    ax = fig.add_subplot(111, projection='3d')
    x = []
    y = []
    z = []
    colors = []
    #Guardamos los valores de los individuos
    for individuo in entrenamiento:
        x.append(individuo[0])
        y.append(individuo[1])
        z.append(individuo[5])
        colors.append(colores.get(individuo[caracteristicas], 'black'))
    
    #Creamos el gráfico 3D, usamos los 2 primeros atributos y el atributo 17
    ax.scatter3D(x, y, z, c=colors, s=30)

    ax.set_xlabel('Atributo 1')
    ax.set_ylabel('Atributo 2')
    ax.set_zlabel('Atributo 5')
    #Guardar figura
    plt.savefig('inicio'+str(iteracion)+'.png')
    #Mostrar la figura
    plt.show() 
    
    
    #Para el gráfico 2D, se toman en cuenta los  primeros 2 atributos
    for num, color in enumerate(colors):
        plt.scatter(x[num], y[num], c=color, edgecolors='none', s=30)
        
    plt.xlabel('Atributo 1')
    plt.ylabel('Atributo 2')
    #Guardar figura
    plt.savefig('fin'+str(iteracion)+'.png')
    #Mostrar figura
    plt.show()
    
```

## Pseudocódigo de k-NN


```python
#Clasificación con k-NN

def vecinos_mas_cercanos(nuevo_individuo, vecinos, distancia, caracteristicas, entrenamiento):
    #Se crea la lista de distancias
    distancias = []
    for individuo in entrenamiento:
        distancias.append([individuo[caracteristicas], funciones(individuo, nuevo_individuo, caracteristicas, distancia)])
    return(votacion(distancias, vecinos))

# Para ordenar la lista de acuerdo a las distancias y no al indice del
def segundo_elemento(elementos):
    return elementos[1]

#Votación
def votacion(distancias, vecinos):
    #Ordenamos las distancias de forma ascendente
    distancias.sort(key=segundo_elemento) 
    
    #Se guardan las clases de los vecinos más cercanos
    clases = []
    for index in range(vecinos):
        clases.append(int(distancias[index][0])) 
    
    #Buscamos la clase con mayor aparición
    counter = 0
    clase = clases[0] 
    for posible_clase in clases: 
        frecuencia = clases.count(posible_clase) 
        if(frecuencia > counter): 
            counter = frecuencia 
            clase = posible_clase
            
    #Retornamos la clase con mayor aparición
    return(clase) 

def confusión(matriz_confusion):
    print("\nMatriz de confusión")
    clases_totales.sort()  
    print(" ",*clases_totales, sep=", ")
    for clase, fila in enumerate(matriz_confusion):
        print("{} {}".format(clase, fila))

def knn(distancia, vecinos, k):
    folds = ceil(len(entrenamiento)/k)

    funciones_distancia = {
            0: "Función invalida",
            1: "Manhattan",
            2: "Euclidiana",
            3: "Minkowski" }

    #Se muestra la función de distacia que se está usando
    print("Distancia: ", funciones_distancia.get(distancia, "Minkowski"))

    #Tamano de los atributos del individuo
    caracteristicas = len(entrenamiento[0]) - 1

    #Error
    error = 0

    #Inicializar la matriz de confusión con 0
    matriz_confusion = [0 for _ in range(len(clases_totales))] 
    matriz_confusion = [copy.deepcopy(matriz_confusion) for _ in range(len(clases_totales))]

    if ( k < 2 ): #Si no hay cross-validation
        for nuevo_individuo in prueba:
            #Obtenemos la clase
            clase = (vecinos_mas_cercanos(nuevo_individuo, vecinos, distancia, caracteristicas, entrenamiento))
            #print("La clase es: {}, La clase debería ser {}".format(clase, int(nuevo_individuo[caracteristicas])))
            if (clase != nuevo_individuo[caracteristicas]):
                error += 1 #Suma del error

            #Colocamos en la matriz de confusión la predicción
            matriz_confusion[int(nuevo_individuo[caracteristicas])][int(clase)] += 1

        #Cálculo de la exactitud
        print((100/len(prueba))* (len(prueba)-error))

    #Cross-Validation   
    else:
        exactitud = 0
        for iteracion in range(k):
            error= 0
            tamaño = (len(entrenamiento)%folds) if iteracion == k-1 else folds
            prueba_k = random.choices(entrenamiento, k=tamaño)
            entrenamiento_k = []
            contador = 0
            for individuo in entrenamiento:
                if ( individuo not in prueba_k or contador == len(prueba_k)):
                    entrenamiento_k.append(individuo)
                else:
                    contador+=1
            for nuevo_individuo in prueba_k:
                #Obtenemos la clase
                clase = (vecinos_mas_cercanos(nuevo_individuo, vecinos, distancia, caracteristicas, entrenamiento_k))
                if (clase != nuevo_individuo[caracteristicas]):
                    error += 1 #Suma del error

                #Colocamos en la matriz de confusión la predicción
                matriz_confusion[int(nuevo_individuo[caracteristicas])][int(clase)] += 1

            #Cálculo de la exactitud
            exactitud += ((100/len(prueba_k)) * (len(prueba_k)-error))

        print("Exactitud: ", exactitud/k)

    #Llamando a la función para mostrar la matriz de confusión
    confusión(matriz_confusion)
```

```python
#1 para distancia Manhattan, 2 para distancia Euclidiana y 3 o más para distancia Minkowsky
distancia = 2

#Numero de vecinos a tomar en cuenta
vecinos = 3

#Cross-Validation, si k es menor a 2 se utiliza el conjunto de test
k=5

knn(distancia, vecinos, k)
```

# Pseudocódigo de k-Means


```python
#Clustering k-means

#Se calcula la distancia a cada centroide, a cada individuo se le coloca el grupo al que pertenece.
def creacion_grupos(entrenamiento_2, centroides, caracteristicas):
    for individuo in entrenamiento_2:
        distancias = []
        for centroide in centroides:
            distancias.append(funciones(centroide, individuo, caracteristicas, 2))
            
        #Se asigna el grupo al que pertenece el individuo
        individuo[caracteristicas] = distancias.index(min(distancias)) 

#Se calcula el nuevo centroide con la media de los elementos agrupados en ese centroide
def nuevos_centroides(entrenamiento_2, centroides, caracteristicas):
    #Se recorren los k centroides
    for grupo, centroide in enumerate(centroides): 
        contador = 0
        suma = [0 for _ in range(caracteristicas)] 
        for individuo in entrenamiento_2: #Se recorren todos los individuos
            if ( grupo == individuo[caracteristicas] ): #Si el individuo pertenece al grupo 
                contador += 1
                suma = [suma[i]+individuo[i] for i in range(caracteristicas)] #Sumamos sus valores
        contador = 1 if contador == 0 else contador
        suma= [suma[i]/contador for i in range(caracteristicas)] #Cálculamos la media
        suma.append(grupo) 
        centroides[grupo] = suma.copy() #Guardamos el nuevo centroide
    return centroides
```

```python
#Creamos una copia del conjunto de entrenamiento
entrenamiento_2 = copy.deepcopy(entrenamiento)

#Número de grupo a formar
clusters = 7

#Iteraciones máximas
iteraciones = 100

#Elección aleatoria de los primero K centroides
centroides = random.choices(entrenamiento_2, k=clusters)

#Tamano de los atributos de los individuos
caracteristicas = len(centroides[0]) - 1

#Se coloca a cada centroide el grupo al final de la lista
for grupo, centroide in enumerate(centroides):
    centroide[caracteristicas] = grupo

for iteracion in range(iteraciones):
    creacion_grupos(entrenamiento_2, centroides, caracteristicas)
    #Para visualizar los grupos iniciales
    if( iteracion == 0 ): visualizacion(entrenamiento_2, caracteristicas, iteracion) 
    centroides_viejos = copy.deepcopy(centroides)
    centroides = nuevos_centroides(entrenamiento_2, centroides, caracteristicas)
    
    #Si los centroides no cambian
    if (centroides_viejos == centroides):   
        break
        
#Visualizar los grupos finales
visualizacion(entrenamiento_2, caracteristicas, iteracion)
        
#Se cuenta cuandos individuos tiene cada grupo   
for grupo, centroide in enumerate(centroides):
    contador = 0
    for individuo in entrenamiento_2:
        if ( grupo == individuo[caracteristicas]):
            contador += 1
    print("El Cluster {} tiene {}".format(grupo,contador))
```

```python
#Inicializar la matriz para ver los elementos que pertenecen al grupo
matriz = [0 for _ in range(len(clases_totales))] 
matriz = [copy.deepcopy(matriz) for _ in range(len(clases_totales))]

#Optiene cuantos individuos de los grupos creados están en las clases originales
for indice_i, individuo in enumerate(entrenamiento_2):
    for indice_e, elemento in enumerate(entrenamiento):
        if( individuo[:caracteristicas] == elemento[:caracteristicas] and (indice_i == indice_e )):
            matriz[int(elemento[caracteristicas])][int(individuo[caracteristicas])] +=1
confusión(matriz)
```
